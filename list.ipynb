```python
# Create a list of 20 numbers
numbers = []

print("Enter 20 numbers:")
for i in range(20):
    num = int(input(f"Enter number {i+1}: "))
    numbers.append(num)

print("Original List:", numbers)

remove_num = int(input("Enter the number to remove duplicates of: "))


if remove_num in numbers:
    first_index = numbers.index(remove_num)
    new_list = []
    count = 0
    
    for num in numbers:
        if num == remove_num:
            count += 1
            if count == 1:
                new_list.append(num)
        else:
            new_list.append(num)
    
    print("Updated List:", new_list)
else:
    print("Number not found in list.")
```
